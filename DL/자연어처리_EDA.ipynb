# text EDA

```python
with open('../datasets/shakespeare.txt') as f:
    shakespeare = f.readlines() #     
    
shakespeare
```

```python
# join() :  (' ')  
token_str = ' '.join(shakespeare)
token_str
```

```python
#    

token_str = token_str.lower()
```

```python
#  

token_str =  token_str.replace('\n', '')
```

```python
#  
import re

#     

# [^a-z] :     
# re.sub(  ,   , )

#     
tokens = re.sub('[^a-z]',' ', token_str)

#   split
tokens = tokens.split(' ')
```

```python
tokens
```

```python
#   ('')

# filter :     
# filter(, ())
#  None ''   
tokens = list(filter(None, tokens))
tokens
```

```python
#      
from collections import Counter

token_cnt = Counter(tokens)
token_cnt
```

```python
# pandas (  )

import pandas as pd

# token_cnt dict  
# dict -> DF   
# orient='index' : dict key index 
# columns=['freq'] : column 
token_df = pd.DataFrame.from_dict(dict(token_cnt),
                                 orient='index',
                                 columns=['freq'])

# index name 
token_df.index.name= 'token'

#    
# by :   column
token_df.sort_values(by='freq', ascending=False, inplace=True)
token_df
```

```python
#    

import matplotlib.pyplot as plt

#  20 
# kind='barh' :  bar 
ax = token_df.head(20).plot(kind='barh')

#    
ax.invert_yaxis() #  y   

# y  
```

```python
# wordcloud 

# wordcloud (   )
from wordcloud import WordCloud

#   
cloud = WordCloud(background_color='white', max_words=50)

# wordcloud    dict  
#    
visual = cloud.fit_words(token_cnt)

# gca() :    
# x, y     
plt.gca().axes.xaxis.set_visible(False)
plt.gca().axes.yaxis.set_visible(False)

#    
# to_image() :  (  ) matplotlib     
plt.imshow(visual.to_image())
```

## count vetorizer  EDA

```python
#     cv  

from sklearn.feature_extraction.text import CountVectorizer

#    shakespeare list 
cv = CountVectorizer(stop_words='english') #   
cnt = cv.fit_transform(shakespeare)

#    (axis=0) 
cnt.toarray().sum(axis=0)
```

```python
#     
cv.get_feature_names_out() 
```

```python
# dict     

# zip :   
cnt_dict = dict(zip(cv.get_feature_names_out(),
                   cnt.toarray().sum(axis=0)))
cnt_dict
```

```python
#   (   df  )

count_df = pd.DataFrame.from_dict(cnt_dict, orient='index',columns=['freq'])
count_df.index.name = 'token'
count_df.sort_values(by='freq', ascending=False, inplace=True)
count_df
```

```python
#   
ax = count_df.head(20).plot(kind='barh')
ax.invert_yaxis()
```

```python
#   

cloud = WordCloud(background_color='white', max_words=100)
visual = cloud.fit_words(cnt_dict)

plt.axis('off') #    
plt.imshow(visual.to_image())
```

## text_to_word_sequence  EDA

```python
from tensorflow.keras.preprocessing.text import text_to_word_sequence

# text_to_word_sequence    
#        text_to_word_sequence
tokens = text_to_word_sequence(' '.join(shakespeare))

#    
token_cnt = Counter(tokens)

# df 
token_df = pd.DataFrame.from_dict(token_cnt, orient='index', columns=['freq'])
token_df.index.name = 'token'
token_df.sort_values(by='freq', ascending=False, inplace=True)
```

```python
#  
ax = token_df.head(20).plot(kind='barh')
ax.invert_yaxis()
```

```python
#  

cloud = WordCloud(background_color='white', max_words=100)
visual = cloud.fit_words(token_cnt)

plt.axis('off')
plt.imshow(visual.to_image())
```

## nltk( )  EDA

```python
# nltk    

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# nltk.download() : nltk     

#   ,      
txt=' '.join(shakespeare).lower()
txt = re.sub('[^a-z]', ' ', txt)
txt
```

```python
# word_tokenize 

tokens = word_tokenize(txt)
tokens
```

```python
# 
stop_words = list(stopwords.words('english'))
stop_words
```

```python
#  
token_list = list(filter(None,map(lambda x: x if x not in stop_words else None, tokens)))
token_list
```

```python
#   
token_cnt = dict(Counter(token_list))

# 
token_df= pd.DataFrame.from_dict(token_cnt, orient='index', columns=['freq'])
token_df.index.name='token'
token_df.sort_values(by='freq', ascending=False, inplace=True)

#  
ax = token_df.head(20).plot(kind='barh')
ax.invert_yaxis()

# wordcloud 
cloud = WordCloud(background_color='white', max_words=100)
visual = cloud.fit_words(token_cnt)

plt.axis('off')
plt.imshow(visual.to_image())

#      
```
