# text EDA

```python
with open('../datasets/shakespeare.txt') as f:
    shakespeare = f.readlines() # 파일의 모든줄을 읽어 리스트로 반환
    
shakespeare
```

```python
# join() :문자열을 다른문자열과 구분자(' ')를 사용하여 결합
token_str = ' '.join(shakespeare)
token_str
```

```python
# 모든 텍스트 소문자로 바꾸기

token_str = token_str.lower()
```

```python
# 역슬래시 없애기

token_str =  token_str.replace('\n', '')
```

```python
# 정규표현식 모듈
import re

# 문자가 아닌 텍스트 다 날리기

# [^a-z] : 소문자 알파벳이 아닌 모든 문자
# re.sub(문서에서 찾을 문자, 찾은 문자를 변경, 문서)

# 소문자 알파벳이 아닌 문자를 공백처리
tokens = re.sub('[^a-z]',' ', token_str)

# 공백을 기준으로 split
tokens = tokens.split(' ')
```

```python
tokens
```

```python
# 비어있는 문자열 지우기('')

# filter : 조건함수를 적용하여 원소를 걸러내는 역할
# filter(조건함수, 대상(문서))
# 여기서 None은 ''아무것도 없는 문자열을 말함
tokens = list(filter(None, tokens))
tokens
```

```python
# 각 단어의 갯수 확인 해 보기
from collections import Counter

token_cnt = Counter(tokens)
token_cnt
```

```python
# pandas로 정리하기(모델학습을 위해 형식변경)

import pandas as pd

# token_cnt가 dict형식이기 때문에 
# dict -> DF 화 방법을 사용
# orient='index' : dict의 key값을 index로 설정
# columns=['freq'] : column이름 지정
token_df = pd.DataFrame.from_dict(dict(token_cnt),
                                 orient='index',
                                 columns=['freq'])

# index name 설정
token_df.index.name= 'token'

# 빈도 수가 큰값부터 정렬
# by : 정렬 할 column
token_df.sort_values(by='freq', ascending=False, inplace=True)
token_df
```

```python
# 단어 빈도수 그래프 그려보기

import matplotlib.pyplot as plt

# 상위 20개만 그래프화
# kind='barh' : 가로 bar 그래프
ax = token_df.head(20).plot(kind='barh')

# 가장 큰값이 위로 오게함
ax.invert_yaxis() # 그래프의 y축을 뒤집어 역순으로 표현함

# y축이 빈도수를 나타내는축
```

```python
# wordcloud로 표현하기

# wordcloud 불러오기(미설치 시 설치 필요)
from wordcloud import WordCloud

# 기본 베이스 잡아주기
cloud = WordCloud(background_color='white', max_words=50)

# wordcloud 만들때는 데이터프레임이 아닌 dict 형식 입력
# 데이터를 담고 있는 객체만들기
visual = cloud.fit_words(token_cnt)

# gca() : 그래프를 그리는 공간 반환
# x, y축의 눈금과 레이블을 보이지 않게 하기
plt.gca().axes.xaxis.set_visible(False)
plt.gca().axes.yaxis.set_visible(False)

# 그래프가 아닌 이미지로 시각화
# to_image() : 이미지 데이터(데이터를 담은 객체)를 matplotlib에서 인식할 수 있는 형태로 변환
plt.imshow(visual.to_image())
```

## count vetorizer 이용해서 EDA

```python
# 위에서 하나씩 정리하던 것들을 cv가 편하게 해줌

from sklearn.feature_extraction.text import CountVectorizer

# 전처리 하기 전인 shakespeare는 list 형태
cv = CountVectorizer(stop_words='english') # 영어의 불용어 제거
cnt = cv.fit_transform(shakespeare)

# 각 단어의 빈도수 열별(axis=0)로 합산
cnt.toarray().sum(axis=0)
```

```python
# 위의 빈도수에 맞는 해당단어 출력
cv.get_feature_names_out() 
```

```python
# dict로 만들어서 단어와 빈도수 같이 보기

# zip : 두 인자를 묶어줌
cnt_dict = dict(zip(cv.get_feature_names_out(),
                   cnt.toarray().sum(axis=0)))
cnt_dict
```

```python
# 데이터프레임으로 만들기 (상단의 직접전처리 후 df화 작업과 동일)

count_df = pd.DataFrame.from_dict(cnt_dict, orient='index',columns=['freq'])
count_df.index.name = 'token'
count_df.sort_values(by='freq', ascending=False, inplace=True)
count_df
```

```python
# 그래프 화 상동
ax = count_df.head(20).plot(kind='barh')
ax.invert_yaxis()
```

```python
# 이미지 화 상동

cloud = WordCloud(background_color='white', max_words=100)
visual = cloud.fit_words(cnt_dict)

plt.axis('off') # 눈금 및 틀 블라인드
plt.imshow(visual.to_image())
```

## text_to_word_sequence 활용 EDA

```python
from tensorflow.keras.preprocessing.text import text_to_word_sequence

# text_to_word_sequence는 불용어 처리를 지원하지 않음
# 따라서 불용어 제거가 필요하다면 불용어 처리한 후 text_to_word_sequence함수이용
tokens = text_to_word_sequence(' '.join(shakespeare))

# 분류한 단어 갯수 세기
token_cnt = Counter(tokens)

# df화 상동
token_df = pd.DataFrame.from_dict(token_cnt, orient='index', columns=['freq'])
token_df.index.name = 'token'
token_df.sort_values(by='freq', ascending=False, inplace=True)
```

```python
# 그래프화 상동
ax = token_df.head(20).plot(kind='barh')
ax.invert_yaxis()
```

```python
# 이미지화 상동

cloud = WordCloud(background_color='white', max_words=100)
visual = cloud.fit_words(token_cnt)

plt.axis('off')
plt.imshow(visual.to_image())
```

## nltk(자연어처리 모듈) 이용하여 EDA

```python
# nltk 미설치 시 설치 필요

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# nltk.download() : nltk에서 제공하는 데이터 및 자원 다운로드

# 전체 소문자 변경, 공백기준으로 단어 합치기 전처리 맨처음과 상동
txt=' '.join(shakespeare).lower()
txt = re.sub('[^a-z]', ' ', txt)
txt
```

```python
# word_tokenize 처리

tokens = word_tokenize(txt)
tokens
```

```python
#불용어 확인해보기
stop_words = list(stopwords.words('english'))
stop_words
```

```python
# 불용어 제거
token_list = list(filter(None,map(lambda x: x if x not in stop_words else None, tokens)))
token_list
```

```python
# 단어 빈도수 카운트
token_cnt = dict(Counter(token_list))

# 데이터프레임화
token_df= pd.DataFrame.from_dict(token_cnt, orient='index', columns=['freq'])
token_df.index.name='token'
token_df.sort_values(by='freq', ascending=False, inplace=True)

# 빈도수 그래프
ax = token_df.head(20).plot(kind='barh')
ax.invert_yaxis()

# wordcloud 이미지
cloud = WordCloud(background_color='white', max_words=100)
visual = cloud.fit_words(token_cnt)

plt.axis('off')
plt.imshow(visual.to_image())

# 전부 위와 동일하니 굳이 출력하지 않음
```
